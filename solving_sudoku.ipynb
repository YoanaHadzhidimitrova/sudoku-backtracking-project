## Problem Statement

The Sudoku board contains 81 cells that are arranged into 9x9 grid, which have to be filled with numbers. Initially, there are some numbers that are populated already in the board and the remaining numbers should be filled based on some rules and constraints. The numbers that can be used are between 1 and 9 inclusively and must appear exactly once in every column, row and small 3x3 section. We can represent each blank cell as a variable $x_{i,j}$, where $i$ represents row index and $j$ represents column index $$ x_{i,j} \in \{1, 2, 3, 4, 5, 6, 7, 8, 9\} $$

If every cell can independently take number from 1 to 9, the number of possible configurations of the Sudoku board would be $9^{81}$. However, we have to follow the constraints, described above in order to have a valid Sudoku board, so this reduces the possible configurations. We need a systematic search method, that will explore the board and eliminate the invalid options in early stage. Backtracking algorithm is a method for solving problems by incrementally building candidates and abandoning them as soon as they violate some of the rules and constraints.

In this project, Sudoku is formulated as constraint satisfaction problem and we will implement Backtracking algorithm in Python in order to solve it.

## Mathematical Background

### Combinatorics

Combinatorics is a branch of mathematics, that studies different ways of arranging elements and ocunting the number of possible configurations that satisfy specific conditions

Sudoku is considered a combinatorial problem, because it is puzzle game that in which the numbers from 1 to 9 have to be arranged following several rules. In each row, column, 3x3 section we can have $9!$ different combinations which is 362 880 combinations in each section. When solving Sudoku puzzle, the solver must determine the correct values for empty cells while satisfying all puzzle constraints. However, most of the possible combinations are invalid, because they violate Sudoku's rules. As a result, the number valid solutions is small, but the search space that must be explored is still very larged. Due to this, the solver must use systematic search strategy that eliminates invalid possibilities early, such as Backtracking algorithm.

### Constraint Satisfaction Problems

A Constraint Satisfaction Problem (CSP) is a mathematical problem, where the solution must meet diffrent constraints. In CSP the objective is to assign values to variables, which meet specific constraints and lay on specific domains => three components of CSP are variables, domains and constraints.\
**Variables** are the things we need to find values for. In our Sudoku example, variables are all empty cells for which we need to find a suitable number for.\
**Domains** are the specific values that can be assigned for variables. In our Sudoku example, domains are the numbers from 1 to 9 inclusively.\
**Constraints** are the rules which restrict how the variables can be assigned. In our Sudoku example, the restriction is that in every row, column and 3x3 subgrid we must contain the numbers from 1 to 9 exactly once.

The **backtracking algoritm** is a method to systematicaly explore possible solutions in CSPs. It operates by assigning values to empty cells and backtracking if the value violates any rule. The algorithm works with the following steps:

    1. Finds variable and assigns value
    2. It reqursively assigns values to subsequent cells
    3. When the assigned value is not correct, the algorith goes to the previous value and tries with different value.
    4. The process continues either until valid vlaue is assigned or all possible options are tried.
    
The method is widely used due to its simplicity, but it can be inefficient for large sets with many variables.

### Search Space

When solving a Sudoku puzzle, each cell should be assigned a number from 1 to 9, which means that for each cell we can have 9 options as a result. One 9x9 grid contains 81 cells, which means that if we have to assign the numbers independantly without following any rules, we would have $9^{81}$ possible combinations, which represents very large search space. When we create rules and constraints, this large number of possible configurations significantly drops as many of the combinations will be invalid. Because of this large search space, we can use algorithms like backtracking in order to explore the possible variations and eliminate part of them in the early stages.

## Methodology

### Backtracking Algorithm

**Backtracking algorithm** is problem-solving technique that finds solution by trying different possibilities and undoing them if they lead to a dead end. The algorithm is commonly used in puzzle problems and constraint satisfaction problems. When it reaches a dead end, it goes to the previous step and explores new possibilities until valid path is found or all options are exhausted.

Steps which Backtracking algorithm is following, when solving Sudoku puzzle:

    1. Find an empty cell (0)
    2. Attempt digits from 1 to 9
    3. Check row, column and subgrid constraints
    4. If the attempt is not valid, go back to the previous attempt and try different number.
    5. Repeat steps 2-4 until valid solution is found or all possible options are exhausted.

### Board Visualization

The function `print_board()` is used to display the Sudoku grid in a readable format. It prints the board row by row and adds visual separators between the 3×3 subgrids to better represent the structure of a standard Sudoku board. This function is used to show both the initial puzzle and the solved board.

```python
def print_board(board):
    for i in range(len(board)):
        if i % 3 == 0 and i > 0:
            print("---------------------")
        for j in range(len(board[i])):
            if j % 3 == 0 and j > 0:
                print("|", end = " ")
            
            print(board[i][j], end = " ")

            if j == len(board[i]) - 1:
                print()
    
```

### Finding an Empty Cell

The function `find_empty()` searches the Sudoku grid for the first empty cell. Empty cells are represented with the value 0. The function returns the position of the empty cell as a pair of indices (row, column). If no empty cell is found, the function returns `None`, which indicates that the board is already complete.

```python
def find_empty(board):
    for i in range(len(board)):
        for j in range(len(board[i])):
            if board[i][j] == 0:
                return i, j
            else:
                if i == 8 and j == 8:
                    if board[i][j] != 0:
                        return None
```

```python
empty_cell_indexes = find_empty(board)
```

### Constraint Validation

The function `valid()` checks whether a candidate number can be placed in a specific cell without violating Sudoku constraints. The function verifies 
three conditions:

1. The number does not already appear in the same row.
2. The number does not appear in the same column.
3. The number does not appear in the corresponding 3×3 subgrid.

If all constraints are satisfied, the function returns `True`; otherwise it returns `False`.

```python
def valid(board, number, empty_cell_indexes):
    if number < 1 or number > 9:
        return False
    row = empty_cell_indexes[0]
    column = empty_cell_indexes[1]
    subgrid_start_row = (row // 3)*3
    subgrid_start_col = (column // 3)*3
    for r in range(0, 9):
        if board[row][r] == number:
            return False
    for c in range(0, 9):
        if board[c][column] == number:
            return False
    for sr in range(subgrid_start_row, subgrid_start_row + 3):
        for sc in range(subgrid_start_col, subgrid_start_col + 3):
            if board[sr][sc] == number:
                return False
    return True
```

### Backtracking Solver

The function `solve()` implements the recursive backtracking algorithm used to complete the Sudoku puzzle. The algorithm repeatedly finds an empty cell and attempts to place numbers from 1 to 9 in that position.
For each candidate number, the function checks whether the placement satisfies all Sudoku constraints. If the placement is valid, the algorithm proceeds recursively to solve the remaining board. If no valid number leads to a solution, the algorithm backtracks by resetting the cell and trying another value.

```python
def solve(board):
    row_column = find_empty(board)
    if row_column == None:
        return True

    for i in range(1,10):
        if valid(board, i, row_column) == True:
            row = row_column[0]
            column = row_column[1]
            board[row][column] = i
            if solve(board) == True:
                return True
            else: 
                board[row][column] = 0
            
    return False
```

### Algorithm Complexity

The backtracking algorithm explores the search space of possible Sudoku configurations by recursively assigning values to empty cells. In the worst case, each empty cell could take up to 9 possible values, which leads to a theoretical worst-case complexity of approximately O(9^n), where n is the 
number of empty cells.
However, Sudoku constraints significantly reduce the number of valid candidate values, allowing the algorithm to prune large portions of the search space. In practice, this makes backtracking efficient enough to solve typical Sudoku puzzles within a short time.

```python
print("Original board:")
print_board(board)

print(solve(board))

print("Solved board:")
print_board(board)
```

## Discussion and Limitations


The backtracking algorithm provides a simple and reliable approach for solving Sudoku puzzles. By systematically trying possible values for each empty cell and reverting choices that violate constraints, the algorithm guarantees that a valid solution will be found if one exists. The implementation demonstrates how recursive search can effectively solve combinatorial problems with well-defined constraints.

However, the backtracking method may become inefficient for very difficult Sudoku puzzles. In the worst case, the algorithm may need to explore a large number of possible configurations before reaching a valid solution. Although the constraints of Sudoku significantly reduce the search space, the algorithm still relies on trial and error.

Several improvements could make the solver more efficient. For example, heuristics could be used to choose the next empty cell more intelligently, such as selecting the cell with the fewest possible candidate values. Other techniques from constraint satisfaction problems, such as constraint propagation or forward checking, could also reduce the number of explored states and improve performance.
