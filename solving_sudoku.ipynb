# Solving Sudoku puzzles using Backtracking Search and Combinatorial Constraints

## Introduction

Sudoku is a well-known logic-based puzzle that can also be viewed as a combinatorial and algorithmic problem. The objective is to fill a 9×9 grid with digits from 1 to 9 so that each row, column, and 3×3 subgrid contains each digit exactly once. In this project, Sudoku is formulated as a constraint satisfaction problem and solved using a backtracking algorithm implemented in Python. The notebook examines the mathematical structure of the puzzle, explains the solving approach step by step, and evaluates the algorithm on puzzles of different difficulty levels. The project demonstrates how a recreational puzzle can be approached through formal mathematical reasoning and systematic computational methods.

## Problem Statement

Sudoku board contains 81 cells, which are in 9x9 grid. The main goal is to fill every column, every row and every 3x3 section with the numbers from 1 to 9, keeping in mind the constraint that the numbers should be exactly one time in each column, row and section. Each cell can be represented with x[i, j], where i is row index, j is column index and $$ x \epsilon [1, 2, 3, 4, 5, 6, 7, 8, 9] $$
